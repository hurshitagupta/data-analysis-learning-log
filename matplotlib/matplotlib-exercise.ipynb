### Matplotlib Practice

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title('Plotting x and y variable')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

### Subplots in matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,4)
plt.plot(x,y)

plt.subplot(2,2,3)
plt.plot(y,x)

plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

### Types of plots

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

### Saving plots

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
plt.savefig('basicplot.png')
```
